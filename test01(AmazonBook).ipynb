```python
import torch
from model.LightGCN import *
from preprocess.AmazonBook import *
from evaluation import *
pd.options.display.max_rows = 10
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
path = './dataset/amazon-book'
dataset = AmazonBook(path)

# Data(num_nodes=144242, edge_index=[2, 2380730], edge_label_index=[2, 603378])
data = dataset.get()
num_users, num_books = dataset.getNumber()
config = {
    'k': 20,
    'lr': 0.001,
    'epochs': 10,
    'num_layers': 2,
    'batch_size': 8192,
    'embedding_dim': 64,
    'num_users': num_users,
    'num_books': num_books,
    'tuning_type': None,
}
model = LightGCN(
    num_nodes=data.num_nodes,
    embedding_dim=config['embedding_dim'],
    num_layers=config['num_layers'],
).to(device)
```

```python
model = lightgcn_evaluation(model, config, data, device)
# Store the model parameters
torch.save(model.state_dict(), f"pretrain/LightGCN_Amazon_Book_{config['epochs']}_Epochs_Top_{config['k']}.pt")
```
